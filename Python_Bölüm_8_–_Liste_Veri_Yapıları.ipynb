```
# Liste Metotları ve Örnekler
L = [1,4,1,1,2,7,2,5,7,9,11,18]
L.append(21) # Liste sonuna 21 elemanını ekler
L
```

```
print(L.count(1)) # Listede kaç tane 1 var sayar
print(L.count(2)) # Listede kaç tane 1 var sayar
```

```
L.sort() # Liste içeriğini sıralar
L
```

```
L.clear() # Listenin tüm elemanlarını siler ve boş liste döndürür.
L
```
